```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import pairwise_distances
from src.config import cols
```

```python
df = pd.read_excel("data/data_for_regression_GB.xlsx")
```

```python
df.columns[0:30]
```

```python
df = df.fillna(0)
```

```python
df = df[cols]
```

```python
y = df["has_green_bond"]

X = df[["relative_total_assets", "deal_year", "cross_nation", "cross_industry", "IG"]]
X = sm.add_constant(X)

logit = sm.Logit(y, X).fit()

df["pscore"] = logit.predict(X)
```

```python
treated = df[df["has_green_bond"] == 1]
controls = df[df["has_green_bond"] == 0]
```

```python
nn = NearestNeighbors(n_neighbors=3)
nn.fit(controls[["pscore"]])

distances, indices = nn.kneighbors(treated[["pscore"]])


matched_controls = controls.iloc[indices.flatten()]



```

```python
matched_sample = pd.concat([treated, matched_controls])
```

```python
matched_sample.to_excel("data/matched_sample.xlsx", index=False)
```

```python
treated = df[df["has_green_bond"] == 1]
controls = df[df["has_green_bond"] == 0]


vars = ["relative_total_assets", "deal_year", "cross_nation", "cross_industry", "IG", "deal_value"]


treated = treated.dropna(subset=vars)
controls = controls.dropna(subset=vars)


X = df[vars].dropna()
cov = np.cov(X.T)
inv_cov = np.linalg.inv(cov)


dist_matrix = pairwise_distances(
    treated[vars],
    controls[vars],
    metric="mahalanobis",
    VI=inv_cov
)

k = 3

matches = []

for i in range(len(treated)):
    idx = np.argsort(dist_matrix[i])[:k]
    matched_controls = controls.iloc[idx]
    matches.append(matched_controls)

matched_controls_df = pd.concat(matches)
matched_sample = pd.concat([treated, matched_controls_df])



```
